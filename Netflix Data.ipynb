```python
df['date_added']=pd.to_datetime(df['date_added'])
```

```python
df['month']=pd.to_datetime(df['date_added'].dt.month,errors='coerce').astype('int64')
```

```python
df['year']=pd.to_datetime(df['date_added'].dt.year,errors='coerce').astype('int64')
```

```python
df.head()
```

```python
df=df.rename(columns={'listed_in':'genre'})
```

```python
df['genre']=df['genre'].apply(lambda x: x.split(",")[0])
```

```python
df.head()
```

Which year has the maximum number of released TV shows and movies?

```python
tv_shows=df[df['type']=='TV Show']
movies=df[df['type']=='Movie']
```

```python
tv_shows['year'].value_counts()
plt.figure(figsize=(10,4))

sns.countplot(x='year',data=df,order=tv_shows['year'].value_counts().index)
plt.title('Most Released TV-Shows in Year')
plt.ylabel('count')
plt.xlabel('Year')

plt.show()
```

```python
movies['year'].value_counts()
plt.figure(figsize=(10,4))

sns.countplot(x='year',data=df,order=movies['year'].value_counts().index)
plt.title('Most Released Movies in Year')
plt.ylabel('count')
plt.xlabel('Year')

plt.show()
```

What is the distribution of content types (TV shows and movies) in the dataset?

```python
df['type'].value_counts()
```

```python
types=df['type'].value_counts()
plt.figure(figsize=(6,4))

plt.pie(types.values, labels=df['type'].value_counts().index, autopct='%1.1f%%',colors=['skyblue','blue'])
plt.title('Distribution of Tv Shows and Movies')
plt.show()
```

Which countries have the highest number of content releases Tv-shows and Movies?

```python
tv_shows['country'].value_counts()
plt.figure(figsize=(8,4))

sns.countplot(x='country',data=df,order=tv_shows['country'].value_counts().iloc[:10].index)
plt.xticks(rotation=90)
plt.title('Countries with Highest No Content Releases')
plt.xlabel('country')
plt.ylabel('count')
```

```python
movies['country'].value_counts()
plt.figure(figsize=(8,4))

sns.countplot(x='country',data=df,order=movies['country'].value_counts().iloc[:10].index)
plt.xticks(rotation=90)
plt.title('Countries with Highest No Content Releases')
plt.xlabel('country')
plt.ylabel('count')
```

List the top 10 directors in the Netflix dataset, considering both TV shows and movies.

```python
df['director'].value_counts()
plt.figure(figsize=(6,5))

sns.countplot(x='director',data=df,order=df['director'].value_counts().iloc[1:10].index)
plt.xticks(rotation=90)
plt.title('Top 10 Directors')
plt.xlabel('Directors')
plt.ylabel('count')
plt.show()
```

How has the distribution of content releases changed over the years?

```python
counts=df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,6))

plt.plot(counts.index,counts.values,marker='*')
plt.title('Distribution of Content Releses Over the Years')
plt.ylabel('No of Releases')
plt.xlabel('Release Year')

plt.show()
```
